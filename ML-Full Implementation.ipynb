```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.predict(x_test)
```

```python
lr.score(x_test,y_test)
```

```python
def adj_r2(x,y):
    r2=lr.score(x,y)
    n=x.shape[0] 
    p=x.shape[1]
    adj_r2=1-((1-r2)*(n-1)/(n-p-1))
    return adj_r2
```

```python
adj_r2(x_test,y_test)
```

```python
#Regularization
from sklearn.linear_model import LassoCV,Lasso,RidgeCV,Ridge,ElasticNetCV,ElasticNet
```

```python
lasso=LassoCV(cv=10,max_iter=20000,normalize=True)
lasso.fit(x_train,y_train)
```

```python
lasso.alpha_
```

```python
lasso=Lasso(alpha=0.0020817816394484118)
lasso.fit(x_train,y_train)
```

```python
ridgecv=RidgeCV(alphas=np.random.uniform(0,10,60),cv=10,normalize=True)
ridgecv.fit(x_train,y_train)
```

```python
ridgecv.alpha_
```

```python
ridge=Ridge(alpha=ridgecv.alpha_)
ridge.fit(x_train,y_train)
```

```python
ridge.score(x_test,y_test)
```

```python
elasticnetcv=ElasticNetCV(cv=10)
```

```python
elasticnetcv.fit(x_train,y_train)
```

```python
elasticnet=ElasticNet(alpha=elasticnetcv.alpha_,l1_ratio=elasticnetcv.l1_ratio_)
```
